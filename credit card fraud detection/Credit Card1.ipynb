##### We can see the Class data is highly skewed. We have to do either upsampling of the minority or downsample the Majority class .Let us further pre process it for now

##### We can also see that except the Amount column, every column has scaled value. Only the Amount column having continuous value which we need to Scale for further preprocessing.

```python
#We scale the amount column and save it in a new column name 'normAmount :
from sklearn.preprocessing import StandardScaler
df['normAmount'] = StandardScaler().fit_transform(np.array(df['Amount']).reshape(-1, 1))

# Let us drop the time and amount column from our analysis for now
df=df.drop(['Time','Amount'],axis=1)
df.head()
```

#### As the dataset is skewed and as mentioned above, we need to resample it.


## Undersampling: 


```python
# In the following lines of code, we undersampled the majority class to the no of minority class.


Number_of_frauds=len(df[df.Class==1])
fraud_indices=np.array(df[df.Class==1].index)
normal_indices=df[df.Class==0].index
random_non_fraud=np.random.choice(normal_indices,Number_of_frauds,replace=False)
random_normal_indices=np.array(random_non_fraud)
under_sample_index=np.concatenate([fraud_indices,random_normal_indices])
under_sample_data=df.iloc[under_sample_index,:]
under_sample_data.head()
```

```python
#lets look at the shape of our new dataframe
print(under_sample_data.shape)
print('the percentage of non fraud cases',len(under_sample_data[under_sample_data.Class==0])/len(under_sample_data))
print('the percentage of  fraud cases',len(under_sample_data[under_sample_data.Class==1])/len(under_sample_data))
```

#### Lets try outlier detection,drop the outliers and see if the correlation gets better.
#### If still no change, we will drop them.

#### To check How to calculate the outlier and remove it, you can see the document here [Outlier removal](https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/)

```python
#outlier removal for V10

q25,q75=np.percentile(v10_fraud_list,25),np.percentile(v10_fraud_list,75)
print('Quartile 25 :{}| Quartile 75: {}'.format(q25,q75))
v10_iqr=q75-q25
print('iqr:{}'.format(v10_iqr))
v10_cutoff=v10_iqr*1.5
v10_lower,v10_upper=q25-v10_cutoff,q75+v10_cutoff
print('Cut_off:{}'.format(v10_cutoff))
print('V10_lower:{}'.format(v10_lower))
print('V10_upper:{}'.format(v10_upper))
print('V10_lower:{}'.format(v10_lower))

print('**' * 44)

outliers=[i for i in v10_fraud_list if i<v10_lower or i>v10_upper]
print('V10_outliers:{}'.format(outliers))
under_sample_data = under_sample_data.drop(under_sample_data[(under_sample_data['V10'] > v10_upper) |
                                                             (under_sample_data['V10'] < v10_lower)].index)
print('Number of Instances after outliers removal: {}'.format(len(under_sample_data)))
print('----' * 44)



```

```python
#outlier removal for V12


q25,q75=np.percentile(v12_fraud_list,25),np.percentile(v12_fraud_list,75)
print('Quartile 25 :{}| Quartile 75: {}'.format(q25,q75))
v12_iqr=q75-q25
print('iqr:{}'.format(v12_iqr))
v12_cutoff=v12_iqr*1.5
v12_lower,v12_upper=q25-v12_cutoff,q75+v12_cutoff
print('Cut_off:{}'.format(v12_cutoff))
print('V12_lower:{}'.format(v12_lower))
print('V12_upper:{}'.format(v12_upper))
print('V12_lower:{}'.format(v12_lower))
print('**' * 44)


outliers=[i for i in v12_fraud_list if i<v12_lower or i>v12_upper]
print('V12_outliers:{}'.format(outliers))
under_sample_data = under_sample_data.drop(under_sample_data[(under_sample_data['V12'] > v12_upper) |
                                                             (under_sample_data['V12'] < v12_lower)].index)
print('Number of Instances after outliers removal: {}'.format(len(under_sample_data)))




```

```python
#outlier removal for V14

q25,q75=np.percentile(v14_fraud_list,25),np.percentile(v14_fraud_list,75)
print('Quartile 25 :{}| Quartile 75: {}'.format(q25,q75))
v14_iqr=q75-q25
print('iqr:{}'.format(v14_iqr))
v14_cutoff=v14_iqr*1.5
v14_lower,v14_upper=q25-v14_cutoff,q75+v14_cutoff
print('Cut_off:{}'.format(v14_cutoff))
print('V14_lower:{}'.format(v14_lower))
print('V14_upper:{}'.format(v14_upper))
print('V14_lower:{}'.format(v14_lower))

print('**' * 44)
outliers=[i for i in v14_fraud_list if i<v14_lower or i>v14_upper]
print('V14_outliers:{}'.format(outliers))
under_sample_data = under_sample_data.drop(under_sample_data[(under_sample_data['V14'] > v14_upper) |
                                                             (under_sample_data['V14'] < v14_lower)].index)
print('Number of Instances after outliers removal: {}'.format(len(under_sample_data)))
```

```python
#lets build our model now
#At first let us separate the feature and target columns

X=under_sample_data.drop('Class',axis=1)
y=under_sample_data['Class']
```

```python

# Classifier Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import collections


# Other Libraries
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline

from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report
from collections import Counter
from sklearn.model_selection import KFold, StratifiedKFold
import warnings
warnings.filterwarnings("ignore")
```

```python
#let us split the data to training and test set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Initialized the classifiers
Classifiers={'LogisticRegression':LogisticRegression(),
            'Knearest':KNeighborsClassifier(),
            'Support Vector':SVC(),
            'DecisionTreeClassifier':DecisionTreeClassifier(),
            'RandoomForestClassifier':RandomForestClassifier()}

```

```python
#we will train the classifiers with cross validation of 5 sets
from sklearn.model_selection import cross_val_score

for key,value in Classifiers.items():
        value.fit(X_train,y_train)
        training_score=cross_val_score(value,X_train,y_train,cv=5)
        print('Classifier:',value.__class__.__name__ ,'has a training score of',round(training_score.mean(),2)*100,'%accuracy score')
```

##### We can see the logistic regression models tops the list with a training accuracy score of 95%
##### Let us fine tune our models again with hyperparameters tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# for logistic regression:
logreg_params={'penalty':['l1','l2'],'C': [0.01,0.10,1,10,100]}
grid_log_reg=GridSearchCV(LogisticRegression(),logreg_params)
grid_log_reg.fit(X_train,y_train)
log_reg=grid_log_reg.best_estimator_
print(log_reg)
```

##### After GridSearchCV we found that the best values for C and penalty is LogReg are 1 nd L1. So lets just build a model with these parameters and predict our testing data

```python

lr=LogisticRegression(C=1,penalty='l1')
lr.fit(X_train,y_train)
y_predlr=lr.predict(X_test)
```

##### Let us try the same for the remaining algorithms

```python
#For Kneighbours
knears_params = {"n_neighbors": list(range(2,5,1)), 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}

grid_knears = GridSearchCV(KNeighborsClassifier(), knears_params)
grid_knears.fit(X_train, y_train)
# KNears best estimator
knears_neighbors = grid_knears.best_estimator_
print(knears_neighbors)
```

```python
kn=KNeighborsClassifier(n_neighbors=2,algorithm='auto')
kn.fit(X_train,y_train)
y_predkn=kn.predict(X_test)
```

```python
# Support Vector Classifier
svc_params = {'C': [0.5, 0.7, 0.9, 1], 'kernel': ['rbf', 'poly', 'sigmoid', 'linear']}
grid_svc = GridSearchCV(SVC(), svc_params)
grid_svc.fit(X_train, y_train)

# SVC best estimator
svc = grid_svc.best_estimator_
print(svc)
```

```python
sv=SVC(C=1,kernel='linear')
sv.fit(X_train,y_train)
y_predsv=sv.predict(X_test)
```

```python

# DecisionTree Classifier
tree_params = {"criterion": ["gini", "entropy"], "max_depth": list(range(2,4,1)), 
              "min_samples_leaf": list(range(5,7,1))}
grid_tree = GridSearchCV(DecisionTreeClassifier(), tree_params)
grid_tree.fit(X_train, y_train)

# tree best estimator
tree_clf = grid_tree.best_estimator_
print(tree_clf)
```

```python
tr=DecisionTreeClassifier(criterion='entropy', max_depth=3,
            max_features=None, max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=5, min_samples_split=2,
            min_weight_fraction_leaf=0.0, presort=False, random_state=None,
            splitter='best')
tr.fit(X_train,y_train)
y_predtr=tr.predict(X_test)
```

```python
#random_forest
Random_param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
grid_rand = GridSearchCV(RandomForestClassifier(), Random_param_grid)
grid_rand.fit(X_train, y_train)

# tree best estimator
grid_rand = grid_rand.best_estimator_
print(grid_rand)
```

```python
rf=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=8, max_features='sqrt', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=200, n_jobs=None,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False)
rf.fit(X_train,y_train)
y_predrf=rf.predict(X_test)
```

```python
#After fitting the models with best hyperparameters, let us check the accuracy of the models

print('The Accuracy of DecisionTree is',accuracy_score(y_test,y_predtr)*100,'%')
print('The Accuracy of Logistic Regression is',accuracy_score(y_test,y_predlr)*100,'%')
print('The Accuracy of Randomforest is',accuracy_score(y_test,y_predrf)*100,'%')
print('The Accuracy of SupportVector is',accuracy_score(y_test,y_predsv)*100,'%')
print('The Accuracy of KNeighbors is',accuracy_score(y_test,y_predkn)*100,'%')
```

```python
#lets 
from sklearn.metrics import roc_curve
from sklearn.model_selection import cross_val_predict
```

```python
log_fpr, log_tpr, log_thresold = roc_curve(y_test, y_predlr)
knear_fpr, knear_tpr, knear_threshold = roc_curve(y_test, y_predkn)
svc_fpr, svc_tpr, svc_threshold = roc_curve(y_test, y_predsv)
tree_fpr, tree_tpr, tree_threshold = roc_curve(y_test, y_predtr)
rf_fpr,rf_tpr,rf_threshold= roc_curve(y_test,y_predrf)

def graph_roc_curve_multiple(log_fpr, log_tpr, knear_fpr, knear_tpr, svc_fpr, svc_tpr, tree_fpr, tree_tpr,rf_fpr,rf_tpr):
    plt.figure(figsize=(16,8))
    plt.title('ROC Curve \n Top 5 Classifiers', fontsize=18)
    plt.plot(log_fpr, log_tpr, label='Logistic Regression Classifier Score: {:.4f}'.format(roc_auc_score(y_test, y_predlr)))
    plt.plot(knear_fpr, knear_tpr, label='KNears Neighbors Classifier Score: {:.4f}'.format(roc_auc_score(y_test, y_predkn)))
    plt.plot(svc_fpr, svc_tpr, label='Support Vector Classifier Score: {:.4f}'.format(roc_auc_score(y_test, y_predsv)))
    plt.plot(tree_fpr, tree_tpr, label='Decision Tree Classifier Score: {:.4f}'.format(roc_auc_score(y_test, y_predtr)))
    plt.plot(rf_fpr,rf_tpr,label='RandomForest Tree Classifier Score: {:.4f}'.format(roc_auc_score(y_test,y_predrf)))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.01, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.annotate('Minimum ROC Score of 50% \n (This is the minimum score to get)', xy=(0.5, 0.5), xytext=(0.6, 0.3),
                arrowprops=dict(facecolor='#6E726D', shrink=0.05),
                )
    plt.legend()
    
graph_roc_curve_multiple(log_fpr, log_tpr, knear_fpr, knear_tpr, svc_fpr, svc_tpr, tree_fpr, tree_tpr,rf_fpr,rf_tpr)



plt.show()
plt.show()
```
